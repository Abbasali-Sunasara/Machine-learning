```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = np.random.rand(5, 5)

plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, cmap='magma', linewidths=0.5)

plt.title("Simple Heatmap Example")
plt.show()
```

```python
categories = ['A','B','A','C','B','A','C','C','B','A']
sns.countplot(x = categories)
plt.title('Count Plot example')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```
