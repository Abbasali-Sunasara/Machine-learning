```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
data = pd.read_csv('files/items.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df.fillna(df.mean(),inplace=True)
df.head()
```

```python
df['Price'].fillna(df['Price'].median(),inplace=True)
df
```

```python
df['Quantity'].fillna(df['Quantity'].std(),inplace=True)
df
```

```python
df['Bought'].fillna(df['Bought'].mean(),inplace=True)
df
```

```python
df['Bought']=df['Bought'].astype(int)
df
```

```python
df['Forenoon'].fillna(df['Forenoon'].max(),inplace=True)
df
```

```python
df['Afternoon'].fillna(method='bfill', inplace=True)
df
```

```python
all_mean = df.fillna(df.mean())
```
