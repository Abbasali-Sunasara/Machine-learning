```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
x = [10,20,30,40,50]
y = [20,25,35,55,45]

plt.plot(x,y)
plt.title("Plotting")
plt.show()
```

```python
x = [10,20,30,40,50]
y = [20,25,35,55,45]

sns.lineplot(x=x,y=y)
plt.title("Plotting")
plt.show()
```

```python
data = pd.read_csv("Files/house_prices.csv")
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
x = df['House Size']
y = df['House Price']
```

```python
plt.scatter(x,y)
plt.show()
```

```python
labels = df['House Price'].value_counts().index  
plt.pie(y, labels=labels)  
plt.title("House Price Distribution")
plt.show()
```
